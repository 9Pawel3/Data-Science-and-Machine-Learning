```python
import numpy as np
import matplotlib.pyplot as plt
```

# `Scatter plot`

```python
x_data = np.random.random(50) * 100 # 50 numbers between 0 and multiplied by 100
y_data = np.random.random(50) * 100

vis = plt.scatter(x_data, y_data, c="r", marker="^", s = 150, alpha=0.3)
plt.show()
```

Line chart

```python
years = [2006 + x for x in range(16)]
weights = [80, 83, 84, 85, 86, 82, 81, 79, 83,
           80, 82, 82, 83, 81, 80, 79]

len(years) == len(weights)

plt.plot(years, weights, c="y", lw = 3, linestyle="--") # line width or plt.plot(years, weights, lw = 3, "y--")
plt.show()
```

Bar plot

```python
prog = ["C++", "C#", "PYthon", "Java", "Go"]
votes = [20, 50, 140, 1, 45]

plt.bar(prog, votes, align="edge", color="r", edgecolor="black", width=0.5, lw=3)
plt.show()
```

Histogram

```python
ages = np.random.normal(20, 1.5, 1000) # list of numbers normal distribution mean 20 years old and standard deviations of 1.5 and 1000 values in total
plt.hist(ages, bins = 20, cumulative=True, color='red')
plt.show()
```

Pie Chart

```python
langs = ['Python', 'C++', 'Java', 'C#', 'Go']
voting = [50, 24, 14, 6, 17]
exp = [0, 0, 0, 0.4, 0]

plt.pie(voting, labels=langs, explode=exp,
        autopct="%.2f##", pctdistance=0.8,
        startangle=90)
plt.show()
```

Customization of plots

```python
years = [2006 + x for x in range(8)]

income = [55, 56, 62, 61,
          72, 72, 73, 75]

income_ticks = list(range(50,81,2)) # values needed in between the income values a little under minimum and above maximum

plt.plot(years, income)
plt.title("Income of John (in USD)")
plt.xlabel = "Year"
plt.ylabel = "Yearly income in USD"
plt.yticks(income_ticks, [f"{x}k USD" for x in income_ticks])
```
