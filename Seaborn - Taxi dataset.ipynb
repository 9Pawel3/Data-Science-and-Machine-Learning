```python
import pandas as pd
import numpy as np

import seaborn as sns
sns.get_dataset_names()

df = sns.load_dataset("taxis")
df
```

```python
df.info()
df.dropna()
```

```python
df['pickup'] = pd.to_datetime(df['pickup']).dt.date
df['dropoff'] = pd.to_datetime(df['dropoff']).dt.date
wykres1 = sns.histplot(df['passengers'], bins=15, kde=True)
```

```python
import matplotlib.pyplot as plt

average_tip_by_payment = df.groupby('payment')['tip'].mean().reset_index()
print(average_tip_by_payment)

sns.barplot(x='payment', y='tip', data=average_tip_by_payment)
plt.xlabel("Payment method")
plt.ylabel("Average tip")
plt.title("Average tip by payment method")
plt.show()
```
